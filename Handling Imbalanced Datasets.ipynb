```python
# Counting the number of each feature in the category
#  From the result, there is a very big difference between the Fraud and non-fraud datasets
df.Class.value_counts()
```

```python
# Separating the dependent and Independent features

X=df.drop('Class', axis=1)
y=df.Class
```

```python
X
```

```python
y
```

```python
print(X.shape)
print(y.shape)
```

```python
# Using diffrent models to obtain results on effects of imbalanced data
# Logistic Regression

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.model_selection import KFold, GridSearchCV
```

```python
log_class=LogisticRegression()
grid={'C':10.0 ** np.arange(-2,3), 'penalty':['l1','l2']}
cv=KFold(n_splits=5,random_state=None,shuffle=False)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.7)
```

```python
clf=GridSearchCV(log_class,grid,cv=cv,n_jobs=1,scoring='f1_macro')
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
#Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier()
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

#### Performing Undersampling
###### Reducing the features with maximum labels to fit the number of minimum labels - Not encouraged coz it leads to loss of data

```python
y_train.value_counts()
```

```python
from collections import Counter
Counter(y_train)
```

```python
from imblearn.under_sampling import NearMiss
ns=NearMiss(sampling_strategy=0.8)
X_train_ns,y_train_ns = ns.fit_resample(X_train,y_train)
print('The number of counts before fit {}'.format(Counter(y_train)))
print('The number of counts after fit {}'.format(Counter(y_train_ns)))
```

```python
#Performing Random Forest Classiefier after balancing the data

from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier()
classifier.fit(X_train_ns,y_train_ns)
```

```python
y_pred=classifier.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

#### CONCLUSION: Introduces many errors, dont use unless the data is very small

### Oversampling

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
over_sampler=RandomOverSampler(sampling_strategy=0.5)
X_train_os,y_train_os = over_sampler.fit_resample(X_train,y_train)
print('The number of counts before fit {}'.format(Counter(y_train)))
print('The number of counts after fit {}'.format(Counter(y_train_os)))
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier()
classifier.fit(X_train_os,y_train_os)
```

```python
y_pred=classifier.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

#### ONCLUSION: Overfitting works better and doesnt need to data loss

```python
from imblearn.combine import SMOTETomek
```

```python
stmek=SMOTETomek(sampling_strategy=0.75)
X_train_st,y_train_st = stmek.fit_resample(X_train,y_train)
print('The number of counts before fit {}'.format(Counter(y_train)))
print('The number of counts after fit {}'.format(Counter(y_train_st)))
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier()
classifier.fit(X_train_st,y_train_st)
```

```python
y_pred=classifier.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

#### Conclusion-This a combination of Over and Under Sampling and equally works well
