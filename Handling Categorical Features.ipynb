### 3. Counter Frequency Encoding - invlove replacing the categorical values with the frequency in which they appear in the dataframe

```python
train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data' , header = None,index_col=None)
train_set.head() 
```

```python
#Picking only the categorical features
columns = [1,3,5,6,7,8,9,13]
train_set = train_set[columns]
```

```python
train_set
```

```python
# Renaming the columns

train_set.rename(columns={
    1:'Employment',
    3:'Education_Level',
    5:'Marital_Status',
    6:'Profession',
    7:'Family_Role',
    8:'Race',
    9:'Gender',
    13:'Country',
},inplace=True)
```

```python
train_set.head()
```

```python
# Finding the number of null values in the dataset

train_set.isnull().sum()
```

```python
# Finding the number of unique labels in each column

for features in train_set.columns:
    print(features , ':', len(train_set[features].unique()), 'labels')
```

```python
# Counting the number of values in Country column

train_set['Country'].value_counts().to_dict()
```

```python
county_map=train_set['Country'].value_counts().to_dict()
```

```python
train_set['Country']=train_set['Country'].map(county_map)
```

```python
train_set.head(20)
```

##### Advantages
##### Easy To Use
##### Not increasing feature space
##### Disadvantages
##### It will provide same weight if the frequencies are same

### 6.Probability Ratio Encoding

#### Find the percentage of survived based on cabin
#### Find the percentage of dead based on cabin ie 1-survived
#### Find the probability ration wich is survived divided by dead and convert to a dictionary
#### Map the dictionary values/probability ratio to each cabin value

```python
# USing the Titanic dataset

df = pd.read_csv('C:/Users/user/DATA_SCIENCE/Feature_Engineering/Datasets/Titanic/titanic.csv', usecols=['Cabin','Survived'])
df.head()
```

```python
# Replacing all missing value with a new feature Missing

df['Cabin'].fillna('Missing', inplace=True)
```

```python
df.Cabin.unique()
```

```python
#Taking the first character in each unique value

df['Cabin']=df['Cabin'].astype(str).str[0]
```

```python
df.head()
```

```python
df.Cabin.unique()
```

```python
# Determining the percentage of survived based on the Cabin column

survival_probability=df.groupby(['Cabin'])['Survived'].mean()
```

```python
survival_probability=pd.DataFrame(survival_probability)
```

```python
survival_probability
```

```python
# Determining the probability of died

survival_probability['Died']=1-survival_probability['Survived']
```

```python
survival_probability.head()
```

```python
survival_probability['Probability_Ratio']=survival_probability['Survived']/survival_probability['Died']
survival_probability.head()
```

```python
# We create a dictionary of the probability ratio and replace the values with the Cabin column

prob_ratio_dict=survival_probability['Probability_Ratio'].to_dict()
prob_ratio_dict
```

```python
df['Carbin_Ratio']=df['Cabin'].map(prob_ratio_dict)
```
